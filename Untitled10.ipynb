```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/US_202402091601.csv',encoding='cp949')
df.head(5)
```

## 데이터 디깅
- 월별 데이터을 기준으로 디깅 하는 방향으로
- 월 별 판매량은 어떤 지 o
- 월 별 어떤 지역이 많이 주문했는지
- 월 별 어떤 카테고리를 많이 주문했느지
- 월 별 순이익은 얼마인지  o


-> 할인을 어떻게 사용할지


```python
df['Date'] = pd.to_datetime(df['formatted_order_date'])
df.head(3)
```

```python
# 총 판매량
df['total_Sales'] = df['Sales']*df['Quantity']
```

```python
# 달
df["Month"] = df["Date"].dt.month

# 요일
df['Day'] = df['Date'].dt.day_name()
df.head(3)
```

```python
df.info()
```

## 월별 총 판매금액
- 8월이 가장 많이 판매됨 -> 이때 주문건수가 많거나 단가가 높은 물품을 샀을 수도
- (x축 1.0,2.0 을 1월/ 2월 이런식으로 바꿀 수는 없는가)


```python
sns.barplot(data=df,x='Month',y='total_Sales')
plt.xticks(fontsize=10)
```

    - 하지만 8월에 총 판매금액 가장 많았는데 주문건수는 높지 않은 것으로 보아 단가가 높은 물품들을 구매해서 그렇지 않을까 예상
    - 9,11,12월은 총 판매금액이 적은데 주문건수가 많을 것을 보아 단가가 낮은 물품을 구매해서 그렇지 않을까 예상

## 총 주문금액이 많은 상위 10개의 지역
- city별 총 판매량을 구하여 상위 10개를 추출
- New York이 압도적으로 많은 것으로 보아 그 주변에 회사가 많을 것으로 추측

- 워싱턴 주 = 시애틀
- 펜실베이아 주 - 필라델피아
- 캘리포니아 주 - LA, 센프란시스코,
- 조지아 주 - 컬럼버스
- 텍사스 주 - 휴스턴
- 플로리다 주 - 탤러해시 , 잭슨
- 일리노이 주 - 시카고

```python
City_group = df.groupby('City').agg({'total_Sales':'sum'})
City = City_group.nlargest(10,'total_Sales')
City
```

```python
City.plot(kind='bar', stacked=True)
plt.xlabel('City')
plt.ylabel('Total_Sales')
plt.title('City of Total_Sales')
plt.xticks(rotation=90)
plt.show()
```

# 주문 날짜에 따른 누적 매출 및 이익 예상

```python
# Sort Values by Order Date
df_sorted = df.sort_values(by=['formatted_order_date'])

# Add cummulative Sales & Profit
df_sorted['cummulative_sales'] = df_sorted['Sales'].cumsum()
df_sorted['cummulative_profit'] = df_sorted['Profit'].cumsum()

df_sorted.head()
```

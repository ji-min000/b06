```python
df = pd.read_csv("/content/US_202402091601.csv",encoding='ISO-8859-1')
df.head(3)
```

```python
df2 = df[["Customer ID","City","State","Category","Sub-Category"]]
df2.head(3)
```

```python
baskets = df2.groupby("Customer ID")["Sub-Category"].apply(list)
from apyori import apriori
rules = apriori(baskets,min_support=0.1,min_confidence=0.5)

for rule in rules:
  print(rule)
```

```python
baskets = df2.groupby("Customer ID")["Category"].apply(list)
from apyori import apriori
rules = apriori(baskets,min_support=0.1,min_confidence=0.5)

for rule in rules:
  print(rule)
```
